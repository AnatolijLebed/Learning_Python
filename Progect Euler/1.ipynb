If we list all the natural numbers below 10 that are multiples of 3 or 5, we get 3, 5, 6 and 9. The sum of these multiples is 23.

Find the sum of all the multiples of 3 or 5 below 1000.

```python
print(sum([i for i in range(1000) if  (i % 3 or i % 5) == 0 ]))
```

What is the largest prime factor of the number 600851475143 ?


```python
def find_simple_mult(number):
    simple_mult = list()
    i = 1
    n = 0
    n += number
    while n > 1:
        i += 1
        while n % i == 0:
            simple_mult.append(i)
            n /= i
    count_mult = 1
    multipliers = list({*simple_mult})
    for i in reversed(multipliers):
        count_mult *= simple_mult.count(i) + 1
        multipliers.append(int(number / i))

    return max(multipliers)


print(find_simple_mult(13195))
```

```python
def find_prime_factors_V1(n, print_prime):
    import time
    from tqdm import tqdm
    factors = list()
    for i in tqdm(range(n - 1 , 1, - 1)):
        if n % i == 0:
            if print_prime: print('Найден делитель - ', i)
            for j in range(2, i):
                if i % j == 0: break
            else:
                factors.append(i)
                if print_prime: print('Найден фактор   - ',i)

    return factors
```
